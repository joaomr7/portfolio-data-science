# Panorama do Covid-19

---

Em dezembro de 2019, a OMS(**Organização Mundial da Saúde**) identificava casos de uma **doença respiratória** na cidade Wuhan, na China. Três meses após, em março de 2020, a OMS já classificava a doença como **pandemia de COVID-19**. Essa doença, científicamente conhecida por **SARS-CoV-2**, contaminou milhões de pessoas inclusive levando à morte de muitas delas.

Esta triste situação levou à paralisação do mundo inteiro. Em vários lugares eram feitos `lockdowns`(confinamentos) para evitar a proliferação do vírus, oque fez com que muitas pessoas começassem a trabalhar remotamente, e não apenas o trabalho mas também a educação, foi transferida para o ambiente remoto.

Contudo, o **objetivo** desse notebook é realizar uma **análise de dados sobre COVID-19**, com a ideia de **obter um panorama sobre a doença**.

## Obtenção dos dados

Os dados a ser utilizado em nossa análise, podem ser obtidos [nesse link](https://github.com/owid/covid-19-data/tree/master/public/data). Basicamente, esses dados são uma coleção de dados mantidos pelo [Our World in Data](https://ourworldindata.org/coronavirus), que possui dados da pandemia de COVID-19 de vários países.

Agora segue uma descrição das variáveis encontradas no conjunto de dados.

**Dicionário de variáveis**

*  **iso_code:** código para nome de países e dependências, conforme a norma `ISO 3166-1`. Alguns valores registrados são do Our World in Data, sendo que esses valores começam com o prefixo `_OWID`.
*  **continent:** continente da localidade geográfica.
*  **location:** localidade geográfica.
*  **date:** data de observação.
*  **total_cases:** total de casos confirmados(incluindo casos prováveis).
*  **new_cases:** novos casos confirmados(incluindo casos prováveis).
*  **new_cases_smoothed:** métrica suavizada de novos casos (incluindo casos prováveis).
*  **total_deaths:** número total de mortes atribuídas ao COVID-19.
*  **new_deaths:** novas mortes atribuídas ao COVID-19.
*  **new_deaths_smoothed:** métrica suavizada de novas mortes atribuídas ao COVID-19.
*  **total_cases_per_million:** total de casos confirmados por milhão de pessoas.
*  **new_cases_per_million:** novos casos confirmados por milhão de pessoas.
*  **new_cases_smoothed_per_million:** métrica suavizada de novos casos confirmados por milhão de pessoas.
*  **total_deaths_per_million:** total de mortes confirmados de COVID-19 por milhão de pessoas.
*  **new_deaths_per_million:** total de novas mortes confirmados de COVID-19 por milhão de pessoas.
*  **new_deaths_smoothed_per_million:** métrica suavizada de novas mortes confirmados de COVID-19 por 1 milhão de pessoas.
*  **reproduction_rate:** taxa de reprodução em tempo real de infecção. Basicamente, um valor alto indica que a doença está se espalhando mais rapidamente, e um valor mais baixo significa que a disseminação está diminuindo.
*  **icu_patients:** número de pacientes em `UTI`(Unidade de Terapia Intensiva), no dia.
*  **icu_patients_per_million:** número de pacientes em `UTI` por milhão de pessoas, no dia.
*  **hosp_patients:** pacientes hospitalizados, no dia.
*  **hosp_patients_per_million:** número de pacientes hospitalizados por milhão de pessoas, no dia.
*  **weekly_icu_admissions:** número de pacientes recém-admitidos nas UTIs, em uma semana(data de observação e o 6 dias anteriores).
*  **weekly_icu_admissions_per_million:** número de pacientes recém-admitidos nas UTIs por milhão de pessoas, em uma semana(data de observação e o 6 dias anteriores).
*  **weekly_hosp_admissions:** número de pacientes recém hospitalizados, em uma semana(data de observação e o 6 dias anteriores).
*  **weekly_hosp_admissions_per_million:** número de pacientes recém hospitalizados por milhão de pessoas, em uma semana(data de observação e o 6 dias anteriores).
*  **total_tests:** total de testes realizados para COVID-19.
*  **new_tests:** novos testes de COVID-19, calculado apenas para dias consecutivos.
*  **total_tests_per_thousand:** total de testes por mil pessoas.
*  **new_tests_per_thousand:** novos teste por mil pessoas.
*  **new_tests_smoothed:** métrica suavizada de novos teste.
*  **new_tests_smoothed_per_thousand:** métrica suavizada de novos teste por mil pessoas.
*  **positive_rate:** porcentagem de testes positivos, dado como uma média contínua de 7 dias.
*  **tests_per_case:** testes realizados por novo caso confirmado de COVID-19.
*  **tests_units:** unidade usada no local para reportar os dados de teste, que podem ser "people tested(pessoas testadas)", "tests performed(testes realizados)" ou "samples tested(amostras testadas)". Importante ressaltar que, uma pessoa pode ser testada mais de uma vez ao dia, e também, mais de uma amostra pode ser requerida para fazer o teste.
*  **total_vaccinations:** total de doses de vacinas administradas.
*  **people_vaccinated:** número de pessoas que receberam ao menos uma dose de vacina.
*  **people_fully_vaccinated:** número de pessoas que receberam todas as doses prescritas pelo protocolo de vacina inicial.
*  **total_boosters:** total de doses de reforço(doses a mais do que as prescritas pelo protocolo de vacina inicial).
*  **new_vaccinations:** novas doses administradas, calculada apenas para dias consecutivos.
*  **new_vaccinations_smoothed:** métrica suavizada de novas doses administradas.
*  **total_vaccinations_per_hundred:** doses de vacina administradas por cem pessoas na população total.
*  **people_vaccinated_per_hundred:** número de pessoas que receberam ao menos uma dose de vacina por cem pessoas na população total.
*  **people_fully_vaccinated_per_hundred:** número de pessoas que receberam todas as doses prescritas pelo protocolo de vacina inicial por cem pessoas na população total.
*  **total_boosters_per_hundred:** total de doses de reforço por cem pessoas na população total.
*  **new_vaccinations_smoothed_per_million:** métrica suavizada de novas doses administradas por milhão de pessoas na população total.
*  **new_people_vaccinated_smoothed:** métrica suavizada do número diário de novas pessoas recebendo a primeira dose da vacina.
*  **new_people_vaccinated_smoothed_per_hundred:** métrica suavizada do número diário de novas pessoas recebendo a primeira dose da vacina por cem pessoas na população total.
*  **stringency_index:** Índice de Severidade de Resposta do Governo - uma medida composta de 9 indicadores de resposta, como fechamento de escolas e locais de trabalho ou proibição de viagens. Os valores vão de 0 a 100, com 100 significando uma resposta mais rigorosa.
*  **population:** população total.
*  **population_density:** densidade demográfica da população(número de pessoas por km² de área).
*  **median_age:** idade média da população.
*  **aged_65_older:** porcentagem da população com 65 anos ou mais.
*  **aged_70_older:** porcentagem da população com 70 anos ou mais.
*  **gdp_per_capita:** PIB(Produto interno bruto) per capita.
*  **extreme_poverty:** porcentagem da população em extrema pobreza.
*  **cardiovasc_death_rate:** taxa de mortalidade por doenças cardiovasculares em 2017 (número anual de mortes por 100.000 pessoas).
*  **diabetes_prevalence:** prevalência de diabetes (porcentagem da população de 20 a 79 anos) em 2017.
*  **female_smokers:** porcentagem de mulheres que fumam.
*  **male_smokers:**  porcentagem de homens que fumam.
*  **handwashing_facilities:** porcentagem da população com instalações básicas de lavagem das mãos nas instalações.
*  **hospital_beds_per_thousand:** camas de hospital por 1000 pessoas.
*  **life_expectancy:** expectativa de vida.
*  **human_development_index:** um índice composto que mede o desempenho médio em três dimensões básicas do desenvolvimento humano - uma vida longa e saudável, conhecimento e um padrão de vida decente.
*  **excess_mortality:** diferença percentual entre o número relatado de mortes semanais ou mensais em 2020-2021 e o número projetado de mortes para o mesmo período com base nos anos anteriores.
*  **excess_mortality_cumulative:** diferença percentual entre o número acumulado de mortes desde 1º de janeiro de 2020 e as mortes projetadas cumulativas para o mesmo período com base nos anos anteriores.
*  **excess_mortality_cumulative_absolute:** diferença cumulativa entre o número relatado de mortes desde 1º de janeiro de 2020 e o número projetado de mortes para o mesmo período com base nos anos anteriores.
*  **excess_mortality_cumulative_per_million:** diferença cumulativa entre o número relatado de mortes desde 1º de janeiro de 2020 e o número projetado de mortes para o mesmo período com base nos anos anteriores, por milhão de pessoas.

**Obs:** `as métricas suavizadas, são uma técnica para reduzir o impacto de flutuações diárias, dessa forma reduzindo ruído nos dados e tornando as tendências mais claras. Todas as métricas suavizadas nesse conjunto de dados foram suavizadas em 7 dias, contando do dia da observação.`

### Evolução do número de casos

Analisando a evolução do número de casos totais, podemos observar como ocorreu o desenvolvimento da pandemia ao passar do tempo. Ainda podemos notar se houve estabilização da situação ou não, ou se há indícios de uma nova onda.

Para isso, primeiramente separaremos os dados que iremos usar para fazer as visualizações de dados. A saber, iremos separar em três grupos as plotagens: países, continentes e grupos de renda. Dessa forma conseguimos ter observações de mais de uma perspectivas.

Por questão de praticidade, vamos primeiramente criar algumas funções para nos auxiliar a separar os dados e facilitar a criação dos gráficos.

```python
# função para pegar os valores requeridos do nosso data frame
def get_values_from_data_frame(df, values):
  # agrupar todos os valores por localidade em diferentes momentos no tempo
  all_locations = df.pivot_table(index='date', columns='iso_code', values=values)

  # filtrar apenas as localidades que se referem a países
  only_countries = [country for country in all_locations.columns if country in iso.countries_by_alpha3]
  only_countries = all_locations[only_countries]
  only_countries.columns = [df[df.iso_code == column].iloc[0].location for column in only_countries.columns]

  # filtrar apenas os continentes
  only_continents = all_locations[['OWID_AFR', 'OWID_ASI', 'OWID_EUR', 'OWID_NAM', 'OWID_OCE', 'OWID_SAM']]
  only_continents.columns = [df[df.iso_code == column].iloc[0].location for column in only_continents.columns]

  # filtrar apenas os grupos de renda
  only_income_groups = all_locations[['OWID_HIC', 'OWID_LIC', 'OWID_LMC', 'OWID_UMC']]
  only_income_groups.columns = [df[df.iso_code == column].iloc[0].location for column in only_income_groups.columns]

  return only_countries, only_continents, only_income_groups
```

```python
# função para plotar dados em gráficos de linha
def plot_line_values(data, title, y_title, cols):
  # configurando o estilo de plot do seaborn
  sns.set_style('ticks')

  # organizar os dados no formato correto
  melted_data = data.reset_index().melt(
    id_vars=['date'],
    value_vars=list(data.columns),
    var_name='variable',
    value_name='value'
  )

  # escalonar os valores para melhor visualização
  melted_data['value'] = melted_data['value'] / 1e6

  # plotar os gráficos principais
  g = sns.relplot(
    data=melted_data,
    x = 'date', y='value', col='variable', hue='variable',
    kind='line', palette='crest', linewidth=4, zorder=5,
    col_wrap=cols, height=2, aspect=1.5, legend=False
  )

  # adicionar título ao plot
  plt.gcf().suptitle(title, fontsize=10)

  # plotar os gráficos de linhas que ficam em menor destaque
  for country, ax in g.axes_dict.items():
    ax.text(.05, .85, country, transform=ax.transAxes)

    sns.lineplot(
      data=melted_data,
       x = 'date', y='value', units='variable',
      estimator=None, color='.8', linewidth=1, ax=ax
    )

    for _, s in ax.spines.items():
      s.set_color('.5')

  # melhorar a visualização do das datas
  ax.set_xticks(ax.get_xticks()[::3])
  date_format = mdates.DateFormatter('%b\n%Y')
  ax.xaxis.set_major_formatter(date_format)

  # adicionar título no eixo y e mostar gráfico
  g.set_titles('')
  g.set_axis_labels('', y_title)
  g.tight_layout()
```

Por fim, podemos começar a plotar os gráficos.

```python
# extrair os dados necessários para plotar os gráficos de total de casos
countries_cases, continents_cases, income_groups_cases = get_values_from_data_frame(df=df, values='total_cases')
```

```python
# selecionar os 10 países com mais casos registrados
countries_with_more_cases = list(countries_cases.iloc[-1].sort_values(ascending=False)[:10].index)
countries_with_more_cases = countries_cases[countries_with_more_cases]

# plotar gráfico do número de casos dos países
plot_line_values(
    data=countries_with_more_cases,
    title='Casos de COVID-19 por Países',
    y_title='N° Casos(Milhão)',
    cols=5)
```

Acima, vemos a evolução do número total de casos nos 10 países com mais casos registrados. Conseguimos observar que os Estados Unidos e China possuem os maiores registros de casos totais, berando por volta dos 100 milhões de casos. Uma observação muito interessante, é que a China, um pouco antes de Janeiro de 2023, teve um enorme aumento no número de casos, que parece ter se estabilizado nos meses seguintes. Já em relação aos outros 8 países, os gráficos parecem seguir um comportamento similar entre eles, com alguns registrando casos mais cedos, outros mais tarde.

```python
# plotar gráfico do número de casos dos continentes
plot_line_values(
    data=continents_cases,
    title='Casos de COVID-19 por Continente',
    y_title='N° Casos(Milhão)',
    cols=3)
```

Já em relação aos continentes, a Ásia, Europa e América do Norte, registram mais casos que os demais continentes, com a Ásia berando os 300 milhões de casos registrados, a Europa, entre 200 e 300 milhões e a América do Norte, berando os 100 milhões. E os dois continente, que possuem menos casos registrados são a África e Oceânia.

```python
# plotar gráfico do número de casos dos grupos de renda
plot_line_values(
    data=income_groups_cases,
    title='Casos de COVID-19 por Grupos de Renda',
    y_title='N° Casos(Milhão)',
    cols=2)
```

Por fim, em relação aos grupos de renda, que a saber são: High income (Alta renda), Low income (Baixa renda), Lower middle income(Renda média baixa) e Upper middle income (Renda média alta). Notamos que há mais casos registrados em países com renda alta e média alta. Já países com baixa e média baixa renda, os números de casos aparentam mais suavizados.

```python
# alterando o estilo do seaborn para dark com grid
sns.set_style('darkgrid')

# configurando o layout dos plots
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10, 6))

# mostar gráfico do número de novos casos suavizados para melhor visualização
sns.lineplot(data=br_df, x='date', y='new_cases_smoothed', linewidth=1.5, ax=ax0)
ax0.fill_between(x=br_df.date, y1=br_df.new_cases_smoothed, alpha=.8)

# mostrar gráfico do número de novas mortes para melhorar a visualização
sns.lineplot(data=br_df, x='date', y='new_deaths_smoothed', linewidth=1.5, ax=ax1)
ax1.fill_between(x=br_df.date, y1=br_df.new_deaths_smoothed, alpha=.8)

# adicionando título
fig.suptitle('Casos e mortes diárias de COVID-19 no Brasil', fontsize=14)

# configurando as labels dos gráficos
ax0.set_xlabel('')
ax0.set_ylabel('Casos Diários', fontsize=14)
ax0.xaxis.set_visible(False)

ax1.set_xlabel('')
ax1.set_ylabel('Mortes Diárias', fontsize=12)
ax1.set_yticks(ax1.get_yticks()[1::2])

# formatando o formato das datas
date_format = mdates.DateFormatter('%b\n%Y')
ax1.xaxis.set_major_formatter(date_format)
ax1.xaxis.grid(False)

# mostrar os gráficos
plt.tight_layout()
```

Podemos observar nos gráficos acima, que entre Janeiro e Maio de 2022, houve o maior pico de casos diários, passando dos 150.000 casos por dia. Interessante notar que neste mesmo período, mesmo sendo o com mais casos, não é o com mais mortes diárias. O pico de mortes foi maior entre Janeiro e Setembro de 2021, chegando a 3.000 mortes diárias.

`Obs: os gráficos foram feitos usando o número de novos casos suavizados e o número de novas mortes suavizados, para possibilitar uma melhor visualização dos gráficos, já que a versão não suavizada contém muita variação de um dia a outro, dificultando a visualização.`

#### Vacinação
A vacinação é de estrema importância para o combate ao vírus, é importante tomar as vacinas para estar protegido contra o COVID-19, é por causa da vacina que conseguimos voltar às atividades normalmente. Por isso, outro ponto interessante de observar é a questão da vacinação no Brasil.

```python
# ordenar os dados pela data, da maior para a menor
date_sorted_br_df = br_df.sort_values(by='date', ascending=False)

# pegar a primeira linha com dados não nulos dos dados ordenados
first_valid_index = date_sorted_br_df.total_vaccinations.first_valid_index()
valid_vaccination = date_sorted_br_df.loc[first_valid_index]

# criar um data frame com informações de vacinação em uma escala de milhão
vaccination = pd.DataFrame.from_dict({
    'Vacinas administradas' : valid_vaccination.total_vaccinations / 1e6,
    'Pessoas vacinadas' : valid_vaccination.people_vaccinated / 1e6,
    'Completamente vacianadas': valid_vaccination.people_fully_vaccinated / 1e6
}, orient='index', columns=['Quantidade'])
vaccination.reset_index(inplace=True)

# alterando o estilo do seaborn para white com grid
sns.set(style='whitegrid')

# configurando figura do plot
fig, ax = plt.subplots(figsize=(8, 4))

# plotar gráfico de barras
sns.barplot(
    data=vaccination,
    x='index',
    y='Quantidade',
    palette='bright',
    ax=ax)

# configurar o título e as labels
formated_date = valid_vaccination.date.strftime('%d/%m/%Y')
ax.set_title(f'Administração de Vacinas de COVID-19 no Brasil\n{formated_date}')

ax.set_ylabel('Quantidade em Milhões')
ax.set_xlabel('')

# mostrar o gráfico
plt.tight_layout()
```

```python
# mostrar porcentagem de pessoas vacinadas e completamente vacinadas
print(f'Situação de vacinacção no Brasil até {formated_date}:')
print(' * {:.2f}% da população vacinada.'.format(valid_vaccination.people_vaccinated / br_df.population.iloc[-1] * 100))
print(' * {:.2f}% da população completamente vacinada.'.format(valid_vaccination.people_fully_vaccinated / br_df.population.iloc[-1] * 100))
```

Podemos notar no gráfico acima que, quase 500 milhões de vacinas foram administradas no Brasil até 22/03/2023. Sendo que 88% da populção tomaram ao menos uma dose, e 81% tomaram todas as doses previstas inicialmente no protocolo de vacina(não contamos o número de doses de reforço).
